```python
from itertools import combinations

def train_test_dict(df_splitted, n_chunks, key):
    
    """
        args : splitted dataset and number of chunks for either training or testing dataset
        returns a dictionary of train or test dataset 
        e.g.
        create new keys test chunks.
        test1 for comb_test(1, 2)   train10 for comb_train(3, 4, 5)
        test2 for comb_test(1, 3)   train9 for comb_train(2, 4, 5)
        test3 for comb_test(1, 4)   train8 for comb_train(2, 3, 5)
        test4 for comb_test(1, 5)   train7 for comb_train(2, 3, 4)
        test5 for comb_test(2, 3)   train6 for comb_train(1, 4, 5)
        test6 for comb_test(2, 4)   train5 for comb_train(1, 3, 5)
        test7 for comb_test(2, 5)   train4 for comb_train(1, 3, 4)
        test8 for comb_test(3, 4)   train3 for comb_train(1, 2, 5)
        test9 for comb_test(3, 5)   train2 for comb_train(1, 2, 4)
        test10 for comb_test(4, 5)  train1 for comb_train(1, 2, 3)
         
    """
    
    comb = list(combinations(df_splitted, n_chunks)) # itertool combination does not return a list, rather tuples. So we convert it to list 

    train_test_dict={}
    for i in range(0, len(comb)):
        z = pd.DataFrame()              # create a new dataframe
        y = comb[i]                     
        for item in y:
            z = pd.concat([z, item])    # concat items
        train_test_dict["{}{}".format(key,i+1)] = z  
        
    return train_test_dict
    

def dataset_spit(df, n_chunks, n_chunk_train, n_chunk_test):
    
    # split data
    df_splitted = np.array_split(df, n_chunks)
    
    # create train dataset dictionary
    return train_test_dict(df_splitted, n_chunk_train, 'train'), train_test_dict(df_splitted, n_chunk_test, 'test')
```
